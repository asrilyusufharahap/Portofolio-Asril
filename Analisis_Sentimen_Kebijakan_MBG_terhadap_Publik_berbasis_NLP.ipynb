```python
# Install Sastrawi library if not already installed
!pip install Sastrawi

import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# ==========================================
# 1. LOAD DATA
# ==========================================
# Pastikan path file sesuai dengan lokasi file Anda
file_path = '/content/tweets-data/mbg.csv' # Sesuaikan nama file jika perlu
df = pd.read_csv(file_path)
print(f"Data berhasil dimuat: {df.shape[0]} baris.")

# ==========================================
# 2. INISIALISASI SASTRAWI
# ==========================================
print("Menyiapkan Sastrawi...")
# Membuat Stemmer (untuk mengubah kata ke bentuk dasar)
factory_stemmer = StemmerFactory()
stemmer = factory_stemmer.create_stemmer()

# Membuat Stopword Remover (untuk menghapus kata umum seperti: yang, dan, di)
factory_stopword = StopWordRemoverFactory()
stopword_remover = factory_stopword.create_stop_word_remover()

# Tambahan stopwords manual (bahasa gaul/singkatan yang tidak ada di Sastrawi)
stopwords_manual = ['yg', 'ga', 'gak', 'kalo', 'udh', 'bgt', 'dr', 'aja', 'sdh',
                    'tdk', 'tuh', 'utk', 'sih', 'gue', 'lo', 'lu', 'gw', 'nih']

# ==========================================
# 3. PREPROCESSING PIPELINE
# ==========================================
def preprocess_text(text):
    # a. Case Folding (Huruf kecil)
    text = str(text).lower()

    # b. Cleaning (Hapus URL, Mention, Hashtag, Angka, Simbol)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#[A-Za-z0-9_]+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # c. Stopword Removal (Sastrawi)
    text = stopword_remover.remove(text)

    # d. Tokenization & Manual Stopword Removal
    tokens = text.split()
    tokens = [word for word in tokens if word not in stopwords_manual]

    # e. Stemming (PENTING: Proses ini bisa agak lama jika datanya banyak)
    # Kita menggabungkan kembali token menjadi kalimat untuk stemming per kalimat
    text_clean = ' '.join(tokens)
    text_stemmed = stemmer.stem(text_clean)

    return text_stemmed

print("Memulai preprocessing (ini mungkin memakan waktu beberapa saat)...")
df['clean_text'] = df['full_text'].apply(preprocess_text)
print("Preprocessing selesai!")

# ==========================================
# 4. ANALISIS SENTIMEN (LEXICON BASED)
# ==========================================
# Update kata kunci berdasarkan konteks data baru (misal: isu keracunan)
pos_words = {'bagus', 'setuju', 'baik', 'gratis', 'untung', 'bangkit', 'dukung',
             'senang', 'hebat', 'terhormat', 'inovasi', 'bantu', 'digital', 'keren',
             'manfaat', 'positif', 'maju', 'sejahtera', 'terima', 'kasih', 'enak', 'aman'}

neg_words = {'tolak', 'masalah', 'kurang', 'lemah', 'terbatas', 'bakar', 'ancam',
             'rugi', 'buruk', 'sulit', 'mahal', 'kecewa', 'bohong', 'gagal', 'susah',
             'negatif', 'hancur', 'benci', 'jelek', 'keracunan', 'sakit', 'muntah',
             'mencret', 'bahaya', 'takut', 'korban', 'parah', 'musibah'}

def get_sentiment(text):
    score = 0
    words = text.split()
    for word in words:
        if word in pos_words:
            score += 1
        elif word in neg_words:
            score -= 1

    if score > 0:
        return 'Positif'
    elif score < 0:
        return 'Negatif'
    else:
        return 'Netral'

df['sentiment'] = df['clean_text'].apply(get_sentiment)

# ==========================================
# 5. VISUALISASI
# ==========================================

# A. Grafik Batang
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=df, order=['Positif', 'Netral', 'Negatif'], palette='viridis')
plt.title('Distribusi Sentimen (Data Baru)', fontsize=14)
plt.xlabel('Sentimen')
plt.ylabel('Jumlah Tweet')
plt.savefig('grafik_sentimen_baru.png') # Simpan gambar
plt.show()

# B. Word Cloud
all_text = ' '.join(df['clean_text'])
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate(all_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud: Isu Utama', fontsize=14)
plt.savefig('wordcloud_baru.png') # Simpan gambar
plt.show()

# Simpan hasil ke CSV
df.to_csv('hasil_analisis_lengkap_sastrawi.csv', index=False)
print("\nSelesai! Hasil disimpan di 'hasil_analisis_lengkap_sastrawi.csv'")
```
